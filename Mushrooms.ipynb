## Mushroom Poison detection with Artificial Neural Networks

### Introduction
Mushroom is one of the important ingredient in our food that has good nutrients. Most types of mushroom are poisonous (inedible), and because of its importance, we need to identify poisonous from eatable mushrooms.
We use Aritificial Neural Networks classify the posionnous mushrooms form the edible ones

### Problem Statement
The dataset includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family Mushroom drawn from The Audubon Society Field Guide to North American Mushrooms (1981). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like "leaflets three, let it be'' for Poisonous Oak and Ivy.

### Project Description : PEAS
Performance measure : Accuracy of detecting poison
Environment : Damp and moist environment which is ideal mushroom habitat
Sensors : Camera, moisture sensor, barometer
Agent type : Model based reflex agent

### Methodology
Artificial Neural Network

### Proposed solution
Using a camera to click pictures of the mushroom, identiy, classify and label images, and create dataset of said mushrooms.

Testing Dataset reveals accuracy of final data.
Using this Neural Networks.

### Conclusion
Any given mushroom given that it is labelled properly, can be now classified and poisonous or edible by this dataset with 89.996 % accuracy.

```python
X = pd.DataFrame(X)
y = pd.DataFrame(y)

# Define a size for your train set
train_size = int(0.8 * len(X))

# Split your dataset
X_train = X[:train_size]
X_test = X[train_size:]

y_train = y[:train_size]
y_test = y[train_size:]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
ann = tf.keras.models.Sequential()

## Adding the input layer and the first hidden layer
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))

## Adding the second hidden layer
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))

## Adding the output layer
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
ann.compile(optimizer= 'adam', loss= 'binary_crossentropy',metrics= ['accuracy'])
```

```python
ann.fit(X_test, y_test, batch_size = 32, epochs = 100)
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.values.reshape(len(y_test),1)),1))
```

```python
def compute_confusion_matrix(true, pred):
  '''Computes a confusion matrix using numpy for two np.arrays
  true and pred.

  Results are identical (and similar in computation time) to:
    "from sklearn.metrics import confusion_matrix"

  However, this function avoids the dependency on sklearn.'''

  K = len(np.unique(true)) # Number of classes
  result = np.zeros((K, K))

  for i in range(len(true)):
    result[true[i]][pred[i]] += 1

  return result
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
